 the distribution of best_price

```python

plt.title("customer_id")
sns.countplot(x='unit_price', data=dataset)
```

 what was the best price of brand

```python

plt.title("quantity")
dataset.groupby('store')['unit'].count().plot(kind='bar')
plt.xlabel("store")
plt.ylabel("total_price")
```

which brand sold their product in lowest price

```python

plt.title("quantity")
dataset.groupby('store')['unit_price'].count().sort_values(ascending=False).head(10).plot(kind='bar')
plt.ylabel("unit_price")
```

category best price of model

```python

plt.title("total_price",color="darkblue", family="serif",size=20)
dataset.groupby('unit_price')['store'].count().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%')
```

 category memory size for device models

```python

plt.title("total_price",color="g", family="serif",size=20)
dataset.groupby('quantity')['unit_price'].count().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%')
```

** We can see that the brand name of Samsung, Apple and Xiaomi had the most sales**

configure the poopularity of os using boxplot method

```python

sns.boxplot(x='quantity', y='store', data=dataset, palette='flare', width=0.8)
plt.xlabel('quantity')
plt.ylabel('unit')
plt.title('unit_price')
plt.figure(figsize=(12,6))
plt.show()
```

check best selling brand

```python

dataset.groupby('unit_price')['total_price'].count().sort_values(ascending=False).head(20).plot(kind='bar', color='deeppink')
plt.xlabel('store')
plt.ylabel('total_price')
plt.title('quantity')
plt.gcf().set_size_inches(8,5)
plt.show()
```
